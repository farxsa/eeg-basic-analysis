# Basic EEG Analysis with MNE-Python
This notebook shows how to perform simple EEG data analysis using basic Python and MNE.

**Steps:**
- Load EEG data
- Apply filtering
- Plot signal
- Calculate power spectral density

```python
# Import basic libraries
import mne
import matplotlib.pyplot as plt
```

```python
# Load sample EEG data
sample_data_path = mne.datasets.sample.data_path()
raw_file = sample_data_path + '/MEG/sample/sample_audvis_raw.fif'
raw = mne.io.read_raw_fif(raw_file, preload=True)
raw.pick_types(meg=False, eeg=True)
raw.crop(tmax=60)  # use only first 60 seconds for simplicity
```

```python
# Apply band-pass filter
raw.filter(1, 30)
```

```python
# Plot EEG signals
raw.plot(n_channels=10, scalings='auto')
```

```python
# Plot power spectral density (frequency domain)
raw.plot_psd(fmax=50)
```
